# Data Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("Example DataSet-Foodcoded.csv")
```

# Data Cleaning

```python
data['GPA'] = pd.to_numeric(data['GPA'], errors='coerce')
data['weight'] = pd.to_numeric(data['weight'], errors='coerce')

gender_map = {1: 'Female', 2: 'Male'}
breakfast_map = {1: 'Cereal', 2: 'Donut'}
exercise_map = {1: 'Everyday', 2: 'Twice or three times per week', 3: 'Once a week', 4: 'Sometimes', 5: 'Never'}

data['Gender'] = data['Gender'].map(gender_map)
data['breakfast'] = data['breakfast'].map(breakfast_map)
data['exercise'] = data['exercise'].map(exercise_map)
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=data, x='GPA', hue='Gender', kde=True, bins=15, palette='viridis')
plt.title("GPA Distribution by Gender")
plt.xlabel("GPA")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=data, x='exercise', palette='pastel', order=data['exercise'].dropna().unique())
plt.title("Exercise Frequency Distribution")
plt.xlabel("Exercise Frequency")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='breakfast', hue='Gender', palette='cool')
plt.title("Breakfast Preference by Gender")
plt.xlabel("Breakfast Preference")
plt.ylabel("Count")
plt.show()
```
